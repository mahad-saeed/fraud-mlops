```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/creditcard.csv')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
print("Data Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nBasic Statistics:")
df.describe()
```

```python
class_counts = df['Class'].value_counts()
print("Class Distribution:")
print(f"Legitimate (0): {class_counts[0]} ({class_counts[0]/len(df)*100:.2f}%)")
print(f"Fraud (1):      {class_counts[1]} ({class_counts[1]/len(df)*100:.2f}%)")

plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.xticks([0,1], ['Legitimate', 'Fraud'])
plt.savefig('../data/class_distribution.png')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].hist(df['Amount'], bins=50, color='steelblue', edgecolor='black')
axes[0].set_title('Transaction Amount Distribution')
axes[0].set_xlabel('Amount')
axes[0].set_ylabel('Count')

axes[1].hist(df['Time'], bins=50, color='coral', edgecolor='black')
axes[1].set_title('Transaction Time Distribution')
axes[1].set_xlabel('Time (seconds)')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.savefig('../data/amount_time_distribution.png')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

df[df['Class']==0]['Amount'].hist(bins=50, ax=axes[0], color='steelblue')
axes[0].set_title('Legitimate Transaction Amounts')
axes[0].set_xlabel('Amount')

df[df['Class']==1]['Amount'].hist(bins=50, ax=axes[1], color='red')
axes[1].set_title('Fraud Transaction Amounts')
axes[1].set_xlabel('Amount')

plt.tight_layout()
plt.savefig('../data/amount_by_class.png')
plt.show()

print("Legitimate - Mean Amount:", df[df['Class']==0]['Amount'].mean().round(2))
print("Fraud      - Mean Amount:", df[df['Class']==1]['Amount'].mean().round(2))
```

```python
plt.figure(figsize=(16, 10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('../data/correlation_heatmap.png')
plt.show()
```

```python
fraud_corr = corr['Class'].drop('Class').sort_values(key=abs, ascending=False)
print("Top 10 features correlated with fraud:")
print(fraud_corr.head(10))

plt.figure(figsize=(8, 5))
fraud_corr.head(10).plot(kind='bar', color='salmon')
plt.title('Top Features Correlated With Fraud')
plt.ylabel('Correlation')
plt.tight_layout()
plt.savefig('../data/fraud_correlations.png')
plt.show()
```

```python
# This is a preview of how you will simulate drift later
# You will shift the Amount feature distribution to simulate production drift

reference = df[df['Time'] <= df['Time'].median()].copy()
drifted = df[df['Time'] > df['Time'].median()].copy()

# Artificially amplify drift in Amount
drifted['Amount'] = drifted['Amount'] * 2.5 + np.random.normal(0, 30, len(drifted))

plt.figure(figsize=(10, 4))
plt.hist(reference['Amount'], bins=50, alpha=0.5, label='Reference (Training)', color='blue')
plt.hist(drifted['Amount'], bins=50, alpha=0.5, label='Drifted (Production)', color='red')
plt.title('Reference vs Drifted Amount Distribution')
plt.xlabel('Amount')
plt.legend()
plt.tight_layout()
plt.savefig('../data/drift_simulation_preview.png')
plt.show()

print("Reference mean Amount:", reference['Amount'].mean().round(2))
print("Drifted mean Amount:  ", drifted['Amount'].mean().round(2))
```
